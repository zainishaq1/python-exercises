```python
# Q1. Reversed string and every 2nd character [cite: 35, 36, 37]
def q1_string_ops(s):
    return s[::-1], s[::2]
print(f"Q1 Result (AI Course): {q1_string_ops('AI Course')}")

# Q2. Squares of even numbers only using list comprehension [cite: 38]
def q2_squares_even(nums):
    return [x**2 for x in nums if x % 2 == 0]
print(f"Q2 Result ([1, 2, 3, 4, 5, 6]): {q2_squares_even([1, 2, 3, 4, 5, 6])}")

# Q3. Frequency of each character using a dictionary [cite: 39]
def q3_char_freq(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    return freq
print(f"Q3 Result ('hello'): {q3_char_freq('hello')}")

# Q4. Indices of all odd numbers using enumerate() [cite: 40]
def q4_odd_indices(lst):
    return [idx for idx, val in enumerate(lst) if val % 2 != 0]
print(f"Q4 Result ([10, 21, 32, 43]): {q4_odd_indices([10, 21, 32, 43])}")

# Q5. Rotate list right by k positions using slicing [cite: 41]
def q5_rotate(lst, k):
    if not lst: return lst
    k = k % len(lst)
    return lst[-k:] + lst[:-k]
print(f"Q5 Result ([1, 2, 3, 4, 5], 2): {q5_rotate([1, 2, 3, 4, 5], 2)}")

# Q6. Two lists to dictionary using zip() [cite: 42]
def q6_zip_to_dict(l1, l2):
    return dict(zip(l1, l2))
print(f"Q6 Result (['a', 'b'], [1, 2]): {q6_zip_to_dict(['a', 'b'], [1, 2])}")

# Q7. Manual iteration with StopIteration handling using try-except [cite: 43, 44]
def q7_manual_iter(lst):
    it = iter(lst)
    while True:
        try:
            val = next(it)
            print(f"Q7 Manual Print: {val}")
        except StopIteration:
            break
print("Q7 Start:")
q7_manual_iter([10, 20, 30])

# Q8. Flatten a nested list using list comprehension [cite: 45]
def q8_flatten(nested):
    return [item for sublist in nested for item in sublist]
print(f"Q8 Result ([[1, 2], [3, 4]]): {q8_flatten([[1, 2], [3, 4]])}")

# Q9. Check palindrome using slicing [cite: 46]
def q9_is_palindrome(s):
    return s == s[::-1]
print(f"Q9 Result ('radar'): {q9_is_palindrome('radar')}")

# Q10. Second largest number without sorting [cite: 47]
def q10_second_largest(nums):
    if len(nums) < 2: return None
    largest = second = float('-inf')
    for n in nums:
        if n > largest:
            second = largest
            largest = n
        elif n > second and n != largest:
            second = n
    return second if second != float('-inf') else None
print(f"Q10 Result ([10, 20, 4, 45]): {q10_second_largest([10, 20, 4, 45])}")
```
